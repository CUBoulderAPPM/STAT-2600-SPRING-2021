This description says that the numbers -1, -2, -4 and -5 all have a
special meaning for this variable. They denote different types of
missingness. You can recode all of these to `NA`, but you should also
think about whether the different missigness indicators are in some way
informative. (i.e., if someone refuses to answer questions related to
drug use, might this inform us about their income?)

The income distributing is right-skewed like one might expect. However,
as indicated in the question description, the income variable is
*topcoded* at the 2% level. More precisely,

``` r
n.topcoded <- with(nlsy, sum(income == max(income, na.rm = TRUE), na.rm = TRUE))
n.topcoded
```

    ## [1] 121

121 of the incomes are topcoded to the maximum value of 2.3588410^{5},
which is the average value of the top 121 earners. You will want to
think about how to deal with this in your analysis.
